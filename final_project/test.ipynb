```python
import gensim
import pandas as pd
```

```python
dataset = pd.read_csv('data/lyrics-data.csv')

dataset = dataset[dataset['Idiom'] == 'ENGLISH']

dataset = dataset[:100]
```

```python
import string
from nltk.corpus import stopwords

def clean_string(s):
    # make lower
    s = s.lower()

    # remove all punctuation
    s = s.translate(str.maketrans('', '', string.punctuation))

    # remove stopwords from s
    s = [word for word in s.split() if not word in stopwords.words()]
    
    return s

# test_string = """This is an arbitrary string with lots of words and it gets longer and longer and long enough to be a test."""
# test_string2 = """This is another string"""
# test_string3 = """No words are allowed in my string"""

# Add column lyrics_clean
dataset['Lyrics_clean'] = dataset['Lyric'].apply(clean_string)
```

```python

data = [d for d in dataset[dataset['Idiom'] == 'ENGLISH']['Lyric']]

# data to lowercase
data = [d.lower() for d in data]

# remove punctuation in data
import string
data = [d.translate(str.maketrans('', '', string.punctuation)) for d in data]

# remove stopwords in data
from nltk.corpus import stopwords
stopwords = set(stopwords.words('english'))
data = [d.split() for d in data]
data = [[d for d in d if d not in stopwords] for d in data]

data
```

```python
def tagged_document(list_of_list_of_words):
   for i, list_of_words in enumerate(list_of_list_of_words):
      yield gensim.models.doc2vec.TaggedDocument(list_of_words, [i])

data_for_training = list(tagged_document(data))
```

```python

model = gensim.models.doc2vec.Doc2Vec(vector_size=40, min_count=2, epochs=10)
model.build_vocab(data_for_training)
model.train(data_for_training, total_examples=model.corpus_count, epochs=model.epochs)
```

```python
ranks = []
second_ranks = []
for doc_id in range(len(data_for_training[:1000])):
    inferred_vector = model.infer_vector(data_for_training[doc_id].words)
    sims = model.dv.most_similar([inferred_vector], topn=1000)
    rank = [docid for docid, sim in sims].index(doc_id)
    ranks.append(rank)

    second_ranks.append(sims[1])

import collections

counter = collections.Counter(ranks)
print(counter)
```
